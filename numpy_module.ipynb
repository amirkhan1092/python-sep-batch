```python
# numpy array as a matrix operation 

import numpy as np 

import matplotlib.pyplot as plt
arr1 = np.arange(1, 10, .1)

sin_collection = np.sin(arr1)
cos_collection = np.cos(arr1)

plt.plot(arr1, sin_collection, linewidth = 4)
plt.plot(arr1, cos_collection, linewidth = 10)

plt.show()
```

```python
# numpy matrix behaviour 

import numpy as np

arr1 = np.arange(1, 13)
arr2 = np.arange(5, 17)

arr1.shape = 3, 4
arr2.shape = 3, 4

sum = np.add(arr1, arr2)

print(arr1)
print(arr2)
print("\n\n Sum is \n", sum)

# print("element arr1(2, 3):-", arr1[2][3])
# print("element arr2(2, 3):-", arr2[3][1])
```
